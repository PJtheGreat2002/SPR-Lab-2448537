## Lab 6: Dynamic Time Warping (DTW)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Plot both vectors to visualize their patterns.

```python
vector1 = np.array([2, 3, 4, 6, 8, 7, 6, 5, 4, 3, 2])
vector2 = np.array([2, 4, 6, 7, 7, 6, 5, 5, 4, 3, 2, 2, 1])
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(vector1, marker='o', label='Vector 1')
plt.title('Vector 1')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(vector2, marker='o', color='orange', label='Vector 2')
plt.title('Vector 2')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
