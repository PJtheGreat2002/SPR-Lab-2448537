# Lab Exercise 2: Fourier Transform and Frequency Spectrum Analysis of Signals


```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. (a) Generate a basic sinusoidal signal in the time domain  
(For example, sine wave: 5 Hz, sampled at 1000 Hz.)

```python
# Signal properties
fs = 1000          # Sampling frequency (Hz)
f = 5              # Signal frequency (Hz)
duration = 1.0     # Duration in seconds
t = np.arange(0, duration, 1/fs)

# Generate sine wave
x = np.sin(2 * np.pi * f * t)
```

### (b) Plot the time-domain waveform of the signal

```python
plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.title('5 Hz Sine Wave (Time Domain)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum
- The DTFT is computed by taking the Fourier Transform over a finely spaced frequency grid (not just at N points).


```python
# DTFT calculation
N = len(x)
w = np.linspace(-np.pi, np.pi, 4096)
X_dtft = np.array([np.sum(x * np.exp(-1j * w_i * np.arange(N))) for w_i in w])

# Convert digital frequency (rad/sample) to analog frequency (Hz)
freq_dtft = w * fs / (2 * np.pi)

plt.figure(figsize=(10, 4))
plt.plot(freq_dtft, np.abs(X_dtft))
plt.title('DTFT Magnitude Spectrum (Continuous Frequencies)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('|X(e^{jω})|')
plt.grid(True)
plt.tight_layout()
plt.show()
```
